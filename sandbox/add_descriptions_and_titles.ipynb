```python
from json import loads

with open('../columns.aliases.json') as json:
    list_of_aliases = loads(json.read())
```

```python
with open('../pipeline.out/iteration-7-before-launch/datapackage.json') as json:
    package = loads(json.read())
```

```python
from slugify import slugify
```

```python
def get_alias(name):
    try:
        if name in list_of_aliases:
            return name
        else:
            for reference, aliases in list_of_aliases.items():
                if aliases:
                    if name in aliases:
                        return reference
                else:
                    raise NotImplementedError(name)
            else:
                raise ValueError(name)
    except ValueError:
        return get_alias(slugify(name, separator='_'))
```

```python
get_alias('Objeto del Gasto')
```

```python
def rename_columns(package):
    schema = package['resources'][0]['schema']
    dimensions = package['model']['dimensions']
    
    for dimension in dimensions.values():
        dimension['primaryKey'] = list(map(get_alias, dimension['primaryKey']))
            
    return package
```

```python
new_package = rename_columns(package)
new_package['model']
```
